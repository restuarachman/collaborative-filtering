# **Data Exploration**

```python
#melakukan transpose dataset
df_transposed = df.T
df_transposed
```

```python
#mengambil salah satu data yaitu pada column "Restu Aditya Rachman"
df_corr = df_transposed.corr(method='pearson', min_periods=1).loc[["Restu Aditya Rachman"],:]
df_corr.drop("Restu Aditya Rachman", axis=1, inplace=True)
```

```python
#menyimpan data restu kemudian disimpan pada variabel df_restu
df_restu = df_transposed["Restu Aditya Rachman"]
df_transposed.drop("Restu Aditya Rachman", axis=1, inplace=True)
```

```python
#menampilkan dataset yang sudah dibersihkan (tidak terdapat restu aditya rachman)
df_transposed
```

```python
#melihat nilai rata-rata dari masing-masing column
df_transposed.mean()
```

```python
#melihat korelasi antar variable
sim = df_corr.values[0]
sim
```

# **Result**

```python
miu = df_transposed.mean().values
miu
```

```python
r_miu = df_transposed.T["BuahNaga"].values - miu
r_miu
```

```python
sim_r_miu = r_miu*sim
sim_r_miu
```

```python
#didapatkan hasil prediksi
prediksi = np.sum(sim_r_miu)/np.sum(sim)+df_restu.mean()
prediksi
```

```python
sum = 0
for i in sim:
  if i > 0.5:
    sum+=i
sum
```

```python
#didapatkan hasil prediksi menggunakan threshold
prediksi2 = sum/np.sum(sim)+df_restu.mean()
prediksi2
```
